```python
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

from nltk import tokenize
```

```python
analyzer = SentimentIntensityAnalyzer()

df = pd.read_csv('scripts.csv')
```

```python
def tw_munge(text):
    return text.replace('\r\n\t', ' ')
```

```python
text = tw_munge(df['tw'][4])

tokenize.sent_tokenize(text)
```

```python
analyzer.polarity_scores(text)
```

```python

import nltk
nltk.download('punkt')
```

```python
scores = [analyzer.polarity_scores(tw_munge(text)) for text in df['tw']]
```

```python
scores_df = pd.DataFrame(scores)
```
